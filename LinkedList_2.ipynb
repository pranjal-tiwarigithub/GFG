Intersection of two sorted Linked lists

```python
#User function Template for python3

''' structure of node:

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

'''

def findIntersection(head1,head2):
    #return head
    
    curr1 = head1
    curr2 = head2
    
    hashmap = {}
    
    while curr1 != None:
        
        if curr1.data in hashmap:
            hashmap[curr1.data] = hashmap[curr1.data] + 1 
        else:
            hashmap[curr1.data] = 1  
        
        curr1 = curr1.next
        
    ll3 = linkedList()
    
    #print(hashmap)
    
    
    while curr2 != None:
        
        if curr2.data in hashmap:
            
            #print(curr2.data)
            
            #print(curr2.data)
            #print(hashmap[2])
            
            if hashmap[curr2.data] != 0 :
                
                #print(curr.data , end = ' ')  #issue 
                ll3.insert(curr2.data)
                hashmap[curr2.data] = hashmap[curr2.data] - 1
        
        
        curr2 = curr2.next 

    
    return ll3.head 
            
            
    


#{ 
 # Driver Code Starts
#Initial Template for Python 3

class Node:
    def __init__(self,data):
        self.data=data
        self.next=None

class linkedList:
    def __init__(self):
        self.head=None
        self.tail=None
    
    def insert(self,data):
        if self.head is None:
            self.head = Node(data)
            self.tail = self.head
        else:
            self.tail.next = Node(data)
            self.tail = self.tail.next

def printList(head):
    while head:
        print(head.data,end=' ')
        head=head.next
        
if __name__ == '__main__':
    t=int(input())
    for _ in range(t):
        
        n1,n2 = [int(x) for x in input().split()]
        arr1 = [int(x) for x in input().split()]
        ll1 = linkedList()
        for i in arr1:
            ll1.insert(i)
        
        arr2 = [int(x) for x in input().split()]
        ll2 = linkedList()
        for i in arr2:
            ll2.insert(i)
        
        result = findIntersection(ll1.head,ll2.head)
        printList(result)
        print()

# } Driver Code Ends
```
